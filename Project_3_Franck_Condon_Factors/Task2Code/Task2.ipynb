```python
import matplotlib.pyplot as plt
```

```python
x = []
y = []

with open('V3.spec.out') as f:
    for line in f:
        line = f.readline()
#         print(line.split('  '))
        line = line.split('  ')
        y.append(float(line[1]))
        x.append(float(line[0]))

# print(y)

plt.figure()
plt.xlabel('Energy (cm-1)', fontsize=12)
plt.ylabel('FC Density',  fontsize=12)
plt.xlim(-300, 2000)
plt.title("")
plt.plot(x, y, label="FC Profile" )
plt.title("Franck-Condon Profile smoothed by the Lorentzian approximation" , fontsize=12)
plt.legend()
plt.show()
```

```python
x = []
y = []

with open('V3.sticks.out') as f:
    for line in f:
        line = f.readline()
#         print(type(line))
        line = line.split(' ')
        y.append(float(line[1]))
        x.append(float(line[0]))
        

plt.figure()
plt.xlabel('Energy (cm-1)', fontsize=12)
plt.ylabel('FC Density', fontsize=12)
plt.title("")
plt.plot(x, y, marker='o', linestyle='none', label="FC Profile" )
plt.title("Franck-Condon Profile", fontsize=12)
plt.legend()
plt.show()
```
